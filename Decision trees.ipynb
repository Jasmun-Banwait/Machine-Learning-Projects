```python
X = pd.get_dummies(df.drop("species",axis=1),drop_first=True)
```

```python
y = df["species"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier()
```

```python
model.fit(X_train,y_train)
```

```python
base_pred = model.predict(X_test)
```

```python
base_pred
```

```python
from sklearn.metrics import classification_report,plot_confusion_matrix
```

```python
print(classification_report(y_test,base_pred))
```

```python
plot_confusion_matrix(model,X_test,y_test)
plt.show()
```

```python
model.feature_importances_
```

```python
X.columns
```

```python
pd.DataFrame(index=X.columns,data=model.feature_importances_,columns=["importance"]).sort_values("importance")
```

```python
from sklearn.tree import plot_tree
```

```python
plt.figure(figsize=(13,8))
plot_tree(model,feature_names=X.columns)
plt.show()
```

```python
X_train
```

```python
def report_model(model):
    model_preds = model.predict(X_test)
    print(classification_report(y_test,model_preds))
    print("\n")
    plt.figure(figsize=(12,8),dpi=200)
    plot_tree(model,feature_names=X.columns)
    plt.show()
    
```

```python
report_model(model)
```

```python
pruened_tree = DecisionTreeClassifier(max_depth=2)
```

```python
pruened_tree.fit(X_train,y_train)
```

```python
report_model(pruened_tree)
```

```python
max_leaf_tree = DecisionTreeClassifier(max_leaf_nodes=3)
max_leaf_tree.fit(X_train,y_train)
```

```python
report_model(max_leaf_tree)
```

```python
entropy_tree = DecisionTreeClassifier(criterion="entropy")
```

```python
entropy_tree.fit(X_train,y_train)
```

```python
report_model(entropy_tree)
```
