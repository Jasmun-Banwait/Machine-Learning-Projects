```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("sonar.all-data.csv")
```

```python
df.head()
```

```python
sns.heatmap(df.corr())
df.corr()
```

```python
df['Target'] = df['Label'].map({'R':0,'M':1})
```

```python
np.abs(df.corr()["Target"].sort_values(ascending=False))[0:6]
```

```python
X = df.drop(["Label","Target"],axis=1)
y = df["Label"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
```

```python


from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
```

```python
scalar = StandardScaler()
```

```python
KN_classifier = KNeighborsClassifier()
```

```python
operations = [("scalar",scalar), ("KN_classifier",KN_classifier)]
```

```python
from sklearn.pipeline import Pipeline
```

```python
pipe = Pipeline(operations)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
K_values = list(range(1,30))
```

```python
param_grid = {"KN_classifier__n_neighbors":K_values}
```

```python
model_grid = GridSearchCV(pipe,param_grid,scoring = "accuracy",cv=5)
```

```python
model_grid.fit(X_train,y_train)
```

```python
model_grid.best_estimator_.get_params()
```

```python
model_grid.cv_results_["mean_test_score"]
```

```python
sns.lineplot(x=K_values,y=model_grid.cv_results_["mean_test_score"])
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
y_pred = model_grid.predict(X_test)
confusion_matrix(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```
